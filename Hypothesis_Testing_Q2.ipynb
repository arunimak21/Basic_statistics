```python
import pandas as pd
import statsmodels.api as sm
import pylab
from scipy.stats import norm
import scipy.stats as ss
```

```python
q2=pd.read_csv('LabTAT.csv')
q2.rename(columns={'Laboratory 1' : 'lab1', 'Laboratory 2':'lab2', 'Laboratory 3':'lab3', 'Laboratory 4':'lab4'}, inplace=True)
q2
```

```python
#NORMALITY TEST
sm.qqplot(q2.lab1,line='s')
pylab.title('QQ plot for lab 1')

sm.qqplot(q2.lab2,line='s')
pylab.title('QQ plot for lab 2')

sm.qqplot(q2.lab3,line='s')
pylab.title('QQ plot for lab 3')

sm.qqplot(q2.lab4,line='s')
pylab.title('QQ plot for lab 4')
```

```python
#VARIANCE TEST
ss.bartlett(q2.lab1,q2.lab2,q2.lab3,q2.lab4)
```

```python
#ONE WAY ANOVA TEST
ss.f_oneway(q2.lab1,q2.lab2,q2.lab3,q2.lab4)
```
